```python
import pandas as pd

stocks=[
'AARTIIND',
'ABB',
'ABCAPITAL',
'ABFRL',
'ACC',
'ADANIGREEN',
'ADANIPORTS',
'ADANIPOWER',
'ADVENZYMES',
'AEGISCHEM',
'AIAENGINEER',
'AJANTPHARM',
'ALKEM',
'ALLCARGO']

# Create a Pandas Excel writer to save the data into an Excel file
writer = pd.ExcelWriter('nse_stocks_data.xlsx', engine='xlsxwriter')

# Loop through all stocks and fetch their data
for stock in stocks:
    try:
        # Fetch the historical data for the stock
        data = request.get_hist(symbol=stock, exchange='NSE', interval=Interval.min_5, n_bars=2000)
        
        
        
#         data = request.get_hist(symbol=symbole,exchange='NSE',interval=Interval.min_5,n_bars=5000)
        # calculate indicators from ta-lib
        rsi = talib.RSI(data["close"], timeperiod=14)
        ema = talib.EMA(data['close'], timeperiod=20)
        macd, signal, hist = talib.MACD(data['close'], fastperiod=12, slowperiod=26, signalperiod=9)
        sma = talib.SMA(data['close'], timeperiod=20)
        adx = talib.ADX(data['high'], data['low'], data['close'], timeperiod=14)
        bb_upper, bb_middle, bb_lower = talib.BBANDS(data['close'], timeperiod=20, nbdevup=2, nbdevdn=2, matype=0)
        obv = talib.OBV(data['close'], data['volume'])
        avgprice = talib.AVGPRICE(data['open'], data['high'], data['low'], data['close'])
    
        data['rsi'] = rsi        
        data['ema'] = ema
        data['macd'] = macd
        data['signal'] = signal
        data['sma'] = sma
        data['adx'] = adx
        data['bb_upper'] = bb_upper
        data['bb_middle'] = bb_middle
        data['bb_lower'] = bb_lower
        data['obv'] = obv
        data['avgprice'] = avgprice
        
        
        
        
        
        
        
        # Convert the data into a Pandas DataFrame
        df = pd.DataFrame(data)

        # Write the data into a new sheet in the Excel file
        df.to_excel(writer, sheet_name=stock)
        
        print(f"Fetched {stock} data")
    except:
        pass

# Save and close the Excel writer
writer.save()
writer.close()
```
