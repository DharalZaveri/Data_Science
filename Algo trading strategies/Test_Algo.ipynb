```python
# Define a list of stock symbols
stock_symbols = [ 'AARTIIND', 'ABB', 'ABCAPITAL', 'ABFRL', 'ACC', 'ADANIGREEN', 'ADANIPORTS', 'ADANIPOWER', 'ADVENZYMES', 'AEGISCHEM', 'AJANTPHARM', 'ALKEM', 'ALLCARGO', 'AMARAJABAT', 'AMBUJACEM', 'APLLTD', 'APOLLOHOSP', 'APOLLOTYRE', 'ARVIND', 'ASAHIINDIA', 'ASHOKLEY', 'ASIANPAINT', 'ASTERDM', 'ASTRAL', 'ASTRAZEN', 'ASTRON', 'ATUL', 'AUBANK', 'AUROPHARMA', 'AVANTIFEED', 'AXISBANK', 'BAJAJFINSV', 'BAJFINANCE', 'BALKRISIND', 'BALMLAWRIE', 'BALRAMCHIN', 'BANDHANBNK', 'BANKBARODA', 'BBL', 'BEL', 'BEPL', 'BGRENERGY', 'BHARATFORG', 'BHARTIARTL', 'BHEL', 'BIOCON', 'BIRLACORPN', 'BLUEDART', 'BOSCHLTD', 'BPCL', 'BRITANNIA', 'CANFINHOME', 'CANBK', 'CASTROLIND', 'CCL', 'CESC', 'CGCL', 'CGPOWER', 'CHAMBLFERT', 'CHENNPETRO', 'CHOLAFIN', 'CIPLA', 'COALINDIA', 'COLPAL', 'CONCOR', 'COROMANDEL', 'CROMPTON', 'CUB', 'CUBEXTUB','CUMMINSIND', 'CYIENT', ]

writer = pd.ExcelWriter('nse_stocks_data.xlsx', engine='xlsxwriter')



# fetch data and calculate indicators
for symbole in stock_symbols:

    try:
        
        data = request.get_hist(symbol=symbole,exchange='NSE',interval=Interval.daily,n_bars=75)
        # calculate indicators from ta-lib
        rsi = talib.RSI(data["close"], timeperiod=14)
        ema = talib.EMA(data['close'], timeperiod=10)
        macd, signal, hist = talib.MACD(data['close'], fastperiod=12, slowperiod=26, signalperiod=9)
        sma = talib.SMA(data['close'], timeperiod=20)
        adx = talib.ADX(data['high'], data['low'], data['close'], timeperiod=14)
        bb_upper, bb_middle, bb_lower = talib.BBANDS(data['close'], timeperiod=20, nbdevup=2, nbdevdn=2, matype=0)
        obv = talib.OBV(data['close'], data['volume'])
        avgprice = talib.AVGPRICE(data['open'], data['high'], data['low'], data['close'])
    
        data['rsi'] = rsi        
        data['ema'] = ema
        data['macd'] = macd
        data['signal'] = signal
        data['sma'] = sma
        data['adx'] = adx
        data['bb_upper'] = bb_upper
        data['bb_middle'] = bb_middle
        data['bb_lower'] = bb_lower
        data['obv'] = obv
        data['avgprice'] = avgprice
        
        df = pd.DataFrame(data)
        df.to_excel(writer, sheet_name=symbole)
        print(symbole)
        
        
    except:
        pass
writer.save()
writer.close()
```
